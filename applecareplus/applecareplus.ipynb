```python
import pandas as pd
csv_path = r"https://docs.google.com/spreadsheets/d/1EvIFLGpsKvuOgdX6F9MYlGl9rCLjbKHPlE8srBDMwYo/edit?usp=sharing"

# Download file from Google Sheets
url = csv_path.replace('/edit?usp=sharing', '/export?format=csv')
df = pd.read_csv(url)
```

## 애케플 제품은 어떠한 단어들이 주로 쓰일까?

```python
apple_care_plus = df[df['apple_care_plus'] == True]

## 형태소 분석 한번
from konlpy.tag import Kkma

keywords = {}
kkma = Kkma()

for i in range(len(apple_care_plus)):
    nouns = kkma.nouns(apple_care_plus.iloc[i]['title'])
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1


    nounts = kkma.nouns(apple_care_plus.iloc[i]['content'])
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1


# print keywords
for k in sorted(keywords, key=keywords.get, reverse=True):
    print(k, keywords[k])
```

### 너무 다양하고, 오타까지 있어서 메뉴얼 필터링 방식에 대해 인사이트가 떠오르지는 않음.

## 뛰어쓰기를 단위로 한번 더 분석

```python

keywords = {}

for i in range(len(apple_care_plus)):
    nouns = apple_care_plus.iloc[i]['title'].split(' ')
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1

    nounts = apple_care_plus.iloc[i]['content'].split(' ')
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1

# print keywords
for k in sorted(keywords, key=keywords.get, reverse=True):
    print(k, keywords[k])
```

## 가설 1: 애플케어플러스 제품은 위와 같은 단어를 반드시 포함하고 있을 것이다.

```python
import re
apple_care_plus = df[df['apple_care_plus'] == True]

rex = re.compile(r'애플케어플러스|애케플|애캐플|케어|캐어|care')
for i in range(len(apple_care_plus)):
    if rex.search(apple_care_plus.iloc[i]['title'].lower()) is None and rex.search(apple_care_plus.iloc[i]['content'].lower()) is None:
        print(apple_care_plus.iloc[i]['id'])
        print(apple_care_plus.iloc[i]['title'])
        print(apple_care_plus.iloc[i]['content'])
        print('---')
```
