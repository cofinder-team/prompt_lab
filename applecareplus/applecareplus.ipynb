```python
import pandas as pd
csv_path = r"https://docs.google.com/spreadsheets/d/1EvIFLGpsKvuOgdX6F9MYlGl9rCLjbKHPlE8srBDMwYo/edit?usp=sharing"

# Download file from Google Sheets
url = csv_path.replace('/edit?usp=sharing', '/export?format=csv')
df = pd.read_csv(url)
```

## ì• ì¼€í”Œ ì œí’ˆì€ ì–´ë– í•œ ë‹¨ì–´ë“¤ì´ ì£¼ë¡œ ì“°ì¼ê¹Œ?

```python
apple_care_plus = df[df['apple_care_plus'] == True]

## í˜•íƒœì†Œ ë¶„ì„ í•œë²ˆ
from konlpy.tag import Kkma

keywords = {}
kkma = Kkma()

for i in range(len(apple_care_plus)):
    nouns = kkma.nouns(apple_care_plus.iloc[i]['title'])
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1


    nounts = kkma.nouns(apple_care_plus.iloc[i]['content'])
    for noun in nouns:
        if noun in keywords:
            keywords[noun] += 1
        else:
            keywords[noun] = 1


# print keywords
for k in sorted(keywords, key=keywords.get, reverse=True):
    print(k, keywords[k])
```

```python
import os
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

rex = re.compile(r'ì• í”Œì¼€ì–´í”ŒëŸ¬ìŠ¤|ì• ì¼€í”Œ|ì• ìºí”Œ|ì¼€ì–´|ìºì–´|care')
llm = ChatOpenAI(temperature=0, openai_api_key= "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", model_name="gpt-3.5-turbo")
test_prompt = PromptTemplate(
    input_variables = ["title", "content"],
    template= open("./prompts/Exp2-2.txt", "r").read()
)
chain = LLMChain(llm = llm, prompt=test_prompt)

apple_care_plus = df[df['apple_care_plus'] == True]
non_apple_care_plus = df[df['apple_care_plus'] == False & (df['title'].str.contains(rex) | df['content'].str.contains(rex))]

sample = non_apple_care_plus.sample(10)

correct = 0
for i in range(len(sample)):
    title = sample.iloc[i]['title']
    content = sample.iloc[i]['content']
    result = chain.run(title=title, content=content)
    actual = "TRUE" if sample.iloc[i]['apple_care_plus'] else "FALSE"
    if result.upper() != actual:
        print(sample.iloc[i]['id'])
        print(sample.iloc[i]['title'])
        print(sample.iloc[i]['content'])
        print(f"Predict: {result.upper()} / Actual: {df.iloc[i]['apple_care_plus']}")
        print('---')
    else:
        correct += 1

sample = apple_care_plus.sample(10)
for i in range(len(sample)):
    title = sample.iloc[i]['title']
    content = sample.iloc[i]['content']
    result = chain.run(title=title, content=content)
    actual = "TRUE" if sample.iloc[i]['apple_care_plus'] else "FALSE"
    if result.upper() != actual:
        print(sample.iloc[i]['id'])
        print(sample.iloc[i]['title'])
        print(sample.iloc[i]['content'])
        print(f"Predict: {result.upper()} / Actual: {df.iloc[i]['apple_care_plus']}")
        print('---')
    else:
        correct += 1

print(f"Accuracy: {correct/20}")
```
